```python
dataset = []
for i in range(len(texts)) :
  dataset.append(kkma.nouns(re.sub('^가-힣a-zA-Z0-9','',texts[i])))
```

```python
pip install gensim
```

```python
from gensim import corpora, models
import gensim
```

```python
data2 = dataset
data2 = [[y for y in x if not len(y) ==1]
        for x in data2]
dictionary2 = corpora.Dictionary(data2)
corpus2 = [dictionary2.doc2bow(text) for text in data2]
```

```python
print(corpus2)
```

```python
import matplotlib.pyplot as plt
from gensim.models import CoherenceModel
```

```python
import random
random.seed(10)  # set the random seed to a fixed value

coherence_values2= []
for i in range(2,10) :
  ldamodel = gensim.models.ldamodel.LdaModel(corpus2, num_topics=i, id2word=dictionary2)
  coherence_model_lda = CoherenceModel(model=ldamodel, texts = data2, dictionary = dictionary2,topn=5)
  coherence_lda = coherence_model_lda.get_coherence()
  coherence_values2.append(coherence_lda)
```

```python
print(coherence_values2)
```

```python
x=range(2,10)
plt.plot(x, coherence_values2)
plt.xlabel('number of topics')
plt.ylabel('coherence score')
plt.show()
```

```python
perplexity_values2 = []
for i in range(2,10):
  ldamodel = gensim.models.ldamodel.LdaModel(corpus2, num_topics=i, id2word=dictionary2)
  perplexity_values2.append(ldamodel.log_perplexity(corpus2))
```

```python
print(perplexity_values2)
```

```python
x = range(2,10)
plt.plot(x, perplexity_values2)
plt.xlabel('number of topics')
plt.ylabel('perplexity score')
plt.show()
```

```python
stop_words = "대통령 100년 100 문재인 연설문 기조연설 대한민국 한반도 청와대 여러분 기념사 모두발언 신년기자회견"
stop_words=stop_words.split(' ')

a = []
for i in dataset:
    b=[]
    for j in i :
        if len(j)>2 :
            if j not in stop_words:
                b.append(j)
    a.append(b)

data = [[y for y in x if not len(y) ==1]
        for x in a]
dictionary = corpora.Dictionary(data)
corpus = [dictionary.doc2bow(text) for text in data]
```

```python
coherence_values2= []
for i in range(2,10) :
  ldamodel = gensim.models.ldamodel.LdaModel(corpus, num_topics=i, id2word=dictionary, random_state=2)
  coherence_model_lda = CoherenceModel(model=ldamodel, texts = data, dictionary = dictionary,topn=10)
  coherence_lda = coherence_model_lda.get_coherence()
  coherence_values2.append(coherence_lda)
```

```python
x=range(2,10)
plt.plot(x, coherence_values2)
plt.xlabel('number of topics')
plt.ylabel('coherence score')
plt.show()
```
